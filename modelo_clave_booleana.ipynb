# Modelo de Claves Booleanas

## Práctica con NLTK

Este notebook implementa un sistema de búsqueda booleana (AND, OR, NOT) sobre dos conjuntos de documentos:

- **Consigna 1:** Documentos sobre inteligencia artificial y aprendizaje automático.
- **Consigna 2:** Documentos sobre civilizaciones antiguas.

Se construye un índice invertido y se permite realizar consultas para obtener los documentos relevantes.


---

### Importación de librerías

```python
import nltk
from nltk.tokenize import word_tokenize
from collections import defaultdict
import unicodedata

#nltk.download("punkt")
```

---

### Funciones comunes

```python

def normalizar(palabra):
    # Pasar a minúsculas y eliminar acentos
    nfkd = unicodedata.normalize('NFKD', palabra)
    return "".join([c for c in nfkd if not unicodedata.combining(c)]).lower()

def tokenizar(texto):
    tokens = word_tokenize(texto.lower())
    tokens = [normalizar(t) for t in tokens if t.isalpha()]
    return tokens

def construir_indice(docs):
    indice = defaultdict(set)
    for doc_id, contenido in docs.items():
        for palabra in tokenizar(contenido):
            indice[palabra].add(doc_id)
    return indice

def buscar(query, indice):
    partes = query.lower().split()
    partes = [normalizar(p) for p in partes]
    
    if "and" in partes:
        p1, _, p2 = partes
        return indice[p1] & indice[p2]
    
    elif "or" in partes:
        p1, _, p2 = partes
        return indice[p1] | indice[p2]
    
    elif "not" in partes:
        p1, _, p2 = partes
        return indice[p1] - indice[p2]
    
    else:
        return indice[partes[0]]

def ejecutar_busquedas(indice):
    while True:
        consulta = input("Ingrese una consulta booleana (o 'salir' para terminar): ")
        if consulta.lower() == "salir":
            break
        resultado = buscar(consulta, indice)
        if resultado:
            print("📄 Documentos encontrados:", resultado)
        else: 
            print("📄 Documentos no encontrados")
```



---

### Consigna 1 – Documentos de IA

```python
docs1 = {
    "doc1": "La inteligencia artificial está revolucionando la tecnología.",
    "doc2": "El aprendizaje automático es clave en la inteligencia artificial.",
    "doc3": "Procesamiento del lenguaje natural y redes neuronales.",
    "doc4": "Las redes neuronales son fundamentales en deep learning.",
    "doc5": "El futuro de la IA está en el aprendizaje profundo."
}

indice1 = construir_indice(docs1)

print("=== Consigna 1 ===")
ejecutar_busquedas(indice1)
```

-------------

### Consigna 2 – Documentos de civilizaciones antiguas

```python
docs2 = {
    "doc1": "Los egipcios construyeron las pirámides y desarrollaron una escritura jeroglífica.",
    "doc2": "La civilización romana fue una de las más influyentes en la historia occidental.",
    "doc3": "Los mayas eran expertos astrónomos y tenían un avanzado sistema de escritura.",
    "doc4": "La antigua Grecia sentó las bases de la democracia y la filosofía moderna.",
    "doc5": "Los sumerios inventaron la escritura cuneiforme y fundaron las primeras ciudades."
}

indice2 = construir_indice(docs2)

print("=== Consigna 2 ===")
ejecutar_busquedas(indice2)
```
